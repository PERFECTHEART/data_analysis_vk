### T1 [1 балл]. Сделайте калькулятор произвольных выражений, операции следующие: +, -, /, ^, *. скобок нет

```python
def calculate(input_str = "0"):  
    # Удаляем пробелы из строки
    nosp=input_str.replace(" ","")
    res = []
    n = 0
    # Формируем список операндов и действий
    for i in range(len(nosp)):
        if nosp[i] in ("+", "-", "*", "/", "^"):
            res.append(int(nosp[n:i]))
            res.append(nosp[i])
            n = i + 1 # позиция очередного числа в строке
    res.append(int(nosp[n:]))
    # Выполняем действия в приоритетном порядке
    # Выполняем возведение в степень справа налево (1 приоритет)
    for i in range(len(res)-1,0,-1):
        if res[i] == "^":
            res[i-1] = res[i-1]**res[i+1]
            # Удаляем из списка 2 элемента: второй аргумент и символ функции (мат.действие)
            res.pop(i+1)
            res.pop(i)
    # Выполняем умножение и деление (2 приоритет)
    for i in range(len(res)):
        while i < len(res) and res[i] in ("*", "/"):
            if res[i] == "*":
                res[i-1] = res[i-1]*res[i+1]
            else:
                res[i-1] = res[i-1]/res[i+1]
            res.pop(i+1)
            res.pop(i)
    # Выполняем сложение и вычитание (3 приоритет)
    for i in range(len(res)):
        while i < len(res) and res[i] in ("+", "-"):
            if res[i] == "+":
                res[i-1] = res[i-1]+res[i+1]
            else:
                res[i-1] = res[i-1]-res[i+1]
            res.pop(i+1)
            res.pop(i)                         
    return res[0]
```

### T2 [1 балл]. Создайте функцию, которая размножает букву, после которой встретилось число, на это число.
HE2LO -> HELLO (число означает сколько раз встречается буква)

```python
def transform_string(inp_string):
    # Изначально строчка пустая
    res = ""
    n = 0 # счетчик повторов буквы, изначально пустой
    for i in range(len(inp_string)):
        # Идем по строке и ищем число
        if not inp_string[i].isdigit():
            if n == 0:
                # Если счетчика не было, печатаем букву 1 раз
                n = 1
            # Если счетчик был, печатаем букву n раз
            res = res+inp_string[i]*n
            n = 0
        # Если у счетчика больше 1 разряда
        else:
            n = n*10 + int(inp_string[i])
    return res
```

### T3 [1 балл]. Создайте функцию, которая выводит статистик по строке: кол-во гласных, согласных, цифр, строчных букв, прописных букв

```python
def string_stat(inp_string):
    # Ъ и Ь не относятся ни к гласным, ни к согласным
    vowels = "aeiouyаеёиоуыэюя"
    consonants ="bcdfghjklmnpqrstvwzxбвгджзйклмнпрстфхцчшщ"
    vc = 0 # счетчик гласных
    cc = 0 # счетчик согласных
    dig = 0 # счетчик цифр
    upc = 0 # счетчик заглавных букв
    lwc = 0 # счетчик строчных букв
    for i in range(len(inp_string)):
        if vowels.find(inp_string[i].lower()) != -1:
            vc += 1
        if consonants.find(inp_string[i].lower()) != -1:
            cc += 1
        if inp_string[i].isdigit():
            dig += 1
        if inp_string[i].islower():
            lwc += 1
        if inp_string[i].isupper():
            upc += 1
        continue
    return (vc, cc, dig, lwc, upc)
```

### Т4 [1 балл]. Напишите функцию, инвертирующую словарь.
Например, из k - [v_1, v_2] в v_1 - [k_1]

```python
def dict_inverter(input_dict):
    res = dict() # cоздаем пустой словарь
    # Список всех значений
    for aval in input_dict.values():
        # Конкретное значение из общего списка
        for val in aval:
            # Создаем пустой список значений инвертируемого списка
            keys = list()
            # Ищем конкретное значение у всех ключей
            for item in input_dict.items():
                for j in range (len(item[1])):
                    # Заносим ключ в список значений
                    if item[1][j] == val:
                        keys.append(item[0])
            # Если список состоит из 1 элемента, преобразуем его в элемент без списка
            res[val] = keys[0] if len(keys) == 1 else keys
    return res
```

#### Протестируйте ее на некоторой матрице

```python
import numpy as np

# Задаём массив и переменную k
array = np.array([[5.1,10.2,3.0,48.2,20.6],[4.4,5.5,6.6,7.7,8.8],[19.2,1.0,31.7,8.8,21.2],[5.2,17.9,26.5,45.3,11.1]])
k = 2
print("Первоначальная матрица: \n",array,"\n")

# Пишем функцию стандартизации массива
def standardize_data(array):
    rows, columns = array.shape # кортеж
    standArray = np.zeros(shape=(rows, columns))
    tempArray = np.zeros(rows)
    for column in range(columns):
        mean = np.mean(array[:,column])
        std = np.std(array[:,column])
        tempArray = np.empty(0)
        for element in array[:,column]:
            tempArray = np.append(tempArray, ((element - mean) / std))
        standArray[:,column] = tempArray
    return standArray

# Стандартизируем массив
array = standardize_data(array)

# Вычисляем ковариантную матрицу, транспонируя массив
covariance_matrix = np.cov(array.T)

# Вычисляем собственные значения и собственные векторы ковариатной матрицы
eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix)

# Вычисляем объясненную дисперсию для каждого компонента
variance_explained = []
for i in eigen_values:
     variance_explained.append((i/sum(eigen_values))*100)
        
# Вычисляем кумулятивную сумму дисперсий
cumulative_variance_explained = np.cumsum(variance_explained)

# Берём первые k компонентов компонента
projection_matrix = (eigen_vectors.T[:][:k]).T

# Получаем матрицу сниженной размерностиы
array_pca = array.dot(projection_matrix)
print("Матрица сниженной размерности: \n",array_pca,"\n")
```

```python
# Берём стандартные операторы (заданием не запрещено)
from operator import sub, mul

class NashNumPy:
    # Создаем подкласс для выполнения инициализации массива    
    class ndarray:
        def __init__(self):
            self.array = [] # наполнение массива
            self.ndim = 0 # число измерений (список или двумерный массив)
            self.shape = (1,) # размерность массива
        # Метод создания массива, когда на вход передаётся готовый массив    
        def iarray(self,data):
            self.ndim = 2
            # Проверяем, существует ли у массива второе измерение
            try:
                i = len(data[0])
            except TypeError:
                self.ndim = 1
            self.array = data
            if self.ndim == 1:
                self.shape = (len(data),)
            else:
                self.shape = ( len(data), len(data[0]))
        # Метод для заполнения массива нулями    
        def izeros(self,size):
            if isinstance(size, list):
                self.ndim = 2
                self.array = [[0 for j in range(size[1])] for i in range(size[0])]
                self.shape = (size[0],size[1])
            else:
                self.ndim = 1
                self.array = [0] * size
                self.shape = (size,)
        # Метод для заполнения массива единицами        
        def iones(self,size):
            if isinstance(size, list):
                self.ndim = 2
                self.array = [[1 for j in range(size[1])] for i in range(size[0])]
                self.shape = (size[0],size[1])
            else:
                self.ndim = 1
                self.array = [1] * size
                self.shape = (size,)
        # Убираем запятые и выравниванием элементы массива по самому длинному элементу
        def __str__(self):
            s="" # Результирующая строка
            lmax = 0 # Самый длинный элемент массива
            if self.ndim == 2:
                # Ищем масксимум и по строчкам и по столбцам
                for i in range(len(self.array)): 
                    for j in range(len(self.array[0])):
                        if lmax < len(str(self.array[i][j])):
                            lmax = len(str(self.array[i][j]))
                # Формируем строки массива с двумя измерениями
                for i in range(len(self.array)):
                    s = s + "[" # каждая строка начинается с открывающейся квадратной скобки
                    for j in range(len(self.array[0])):
                        s2 = str(self.array[i][j]) # формируем строковое представление элемента массива [i][j]
                        # Дополняем строку пробелами слева до максимального размера
                        s = s + " " * (lmax - len(s2) + 1) + s2
                    s = s + " ]"
                    # Перевод строки в конце каждой строчки кроме последней
                    if i < (len(self.array))-1:
                        s = s + "\n"
            else:
                # Формируем строку (массив с одним измерением)
                for i in range(len(self.array)):
                    if lmax < len(str(self.array[i])):
                        lmax = len(str(self.array[i]))
                s="["
                for i in range(len(self.array)):
                    s2 = str(self.array[i])
                    s = s + " " * (lmax - len(s2) + 1) + s2
                s = s + " ]"
            return s
        # Применение некоторой (подаваемой на вход) функции ко всем элементам массива     
        def vectorize(self, func):
            if self.ndim == 1:
                self.array = list(map(func, self.array))
            else:
                r = []
                for row in self.array:
                    row = list(map(func, row))
                    r.append( row )
                self.array = r
        # Заменяем оператор сложения        
        def __add__(self,new):
            if self.shape != new.shape:
                    return("Размеры массивов не совпадают")
            if self.ndim == 2:
                res = [[self.array[i][j] + new.array[i][j]  for j in range (len(self.array[0]))] for i in range(len(self.array))] 
            else:
                res = [sum(i) for i in zip(self.array,new.array)]
            a = NashNumPy.ndarray()
            a.iarray(res)
            return a
        # Заменяем оператор вычитания
        def __sub__(self,new):
            if self.shape != new.shape:
                    return("Размеры массивов не совпадают")
            if self.ndim == 2:
                res = [[self.array[i][j] - new.array[i][j] for j in range (len(self.array[0]))] for i in range(len(self.array))] 
            else:
                res = list(map(sub,self.array,new.array))
            a = NashNumPy.ndarray()
            a.iarray( res )
            return a
        # Заменяем оператор умножения
        def __mul__(self,new):
            if len(new.array)!=len(self.array[0]):
                return("Матрицы не могут быть перемножены")
            r1 = len(self.array)    # количество строк в первой матрице
            c1 = len(self.array[0]) # количество столбцов в 1
            r2=c1
            c2=len(new.array[0]) # количество столбцов во 2ой матрице
            res = []
            for z in range(r1):
                t=[]
                for j in range(c2):
                    s = 0
                    for i in range(c1):
                        s = s + self.array[z][i] * new.array[i][j]
                    t.append(s)
                res.append(t)
            a = NashNumPy.ndarray()
            a.iarray(res)
            return a            
        # Произведение Адамара ("/")
        def __truediv__(self,new):
            if self.shape != new.shape:
                return("Матрицы не могут быть перемножены")
            if self.ndim == 2:
                res = [[self.array[i][j] * new.array[i][j] for j in range (len(self.array[0]))] for i in range(len(self.array))] 
            else:
                res = list(map(mul,self.array,new.array))
            a = NashNumPy.ndarray()
            a.iarray(res)
            return a
        # Вертикальное присоединение
        def vstack(self,new):
            r1 = 1 if self.ndim == 1 else self.shape[0]
            c1 = self.shape[0] if self.ndim == 1 else self.shape[1]
            r2 = 1 if new.ndim == 1 else new.shape[0]
            c2 = new.shape[0] if new.ndim == 1 else new.shape[1]
            if c1 != c2:
                return("Несовместимые массивы: не совпадает число столбцов %d и %d") % (c1,c2)
            res = []
            if r1 == 1:
                res.append(self.array)
            else:
                for row in self.array:
                    res.append(row)
            if r2 == 1:
                res.append(new.array)
            else:
                for row in new.array:
                    res.append(row)
            a = NashNumPy.ndarray()
            a.iarray(res)
            return a
        # Горизонтальное присоединение
        def hstack(self,new):
            r1 = 1 if self.ndim == 1 else self.shape[0]
            c1 = self.shape[0] if self.ndim == 1 else self.shape[1]
            r2 = 1 if new.ndim == 1 else new.shape[0]
            c2 = new.shape[0] if new.ndim == 1 else new.shape[1]
            if r1 != r2:
                return("Несовместимые массивы: не совпадает число строк %d и %d") % (r1,r2)
            if r1 == 1:
                res = []
                for i in range(c1):
                    res.append(self.array[i])
                for i in range(c2):
                    res.append(new.array[i])
            else:
                res = list(range(r1))
                for i in range(r1):
                    res[i] = list(range(c1+c2))
                for i in range(r1):
                    for j in range(c1):
                        res[i][j] = self.array[i][j]
                    for j in range(c2):
                        res[i][c1+j] = new.array[i][j]
            a = NashNumPy.ndarray()
            a.iarray(res)
            return a
                
    def array(data):
        a = NashNumPy.ndarray()
        a.iarray(data)
        return a
    
    def zeros(size):
        a = NashNumPy.ndarray()
        a.izeros(size)
        return a
    
    def ones(size):
        a = NashNumPy.ndarray()
        a.iones(size)
        return a
    
def inc(a):
    return a+1
def dec(a):
    return a-1
 
```
